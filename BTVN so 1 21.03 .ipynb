```python
#BT số 1: Vẽ cờ Senegal
#1 nhập đối tượng
import turtle
#2 Tạo đối tượng
flag = turtle.Turtle()
flag.shape('turtle')
flag.pensize(1)
flag.pencolor()
flag.speed(3)
# Điều khiển 
#1. Ô màu đỏ bên phải
flag.pencolor('red')
flag.fillcolor('red')
flag.begin_fill()
flag.penup()
flag.forward(225)
flag.left(90)
flag.pendown()
flag.forward(150)
flag.left(90)
flag.forward(150)
flag.left(90)
flag.forward(300)
flag.left(90)
flag.forward(150)
flag.left(90)
flag.forward(150)
flag.end_fill()
# ô màu vàng ở giữa
flag.penup()
flag.forward(150)
flag.left(90)
flag.forward(150)
flag.pendown()
flag.pencolor('yellow')
flag.fillcolor('yellow')
flag.begin_fill()
flag.forward(150)
flag.left(90)
flag.forward(300)
flag.left(90)
flag.forward(150)
flag.end_fill()
# ô màu xanh lá cây bên trái
flag.pencolor("green")
flag.penup()
flag.backward(150)
flag.pendown()
flag.backward(150)
flag.fillcolor('green')
flag.begin_fill()
flag.right(90)
flag.backward(300)
flag.right(90)
flag.backward(150)
flag.right(90)
flag.backward(300)
flag.end_fill()
# Vẽ ngôi sao chính giữa

flag.forward(300)
flag.right(90)
flag.penup()
flag.forward(75)
flag.right(90)
flag.forward(100)
flag.pendown
flag.fillcolor('green')
flag.begin_fill()
#cạnh 1
flag.right(18)
flag.forward(100)
#cạnh 2
flag.left(144)
flag.forward(100)
#cạnh 3
flag.left(144)
flag.forward(100)
#cạnh 4
flag.left(144)
flag.forward(100)
#cạnh 5
flag.left(144)
flag.forward(100)
flag.end_fill()



#hoàn thành
turtle.done()



```
