```python
#BT số 1: Vẽ cờ Senegal
#1 nhập đối tượng
import turtle
#2 Tạo đối tượng
flag = turtle.Turtle()
flag.shape('turtle')
flag.pensize(1)
flag.pencolor()
flag.speed(3)
# Điều khiển 
#1. Ô màu đỏ bên phải
#khai báo biến
#Chiều dài
x = 720
#Chiều rộng
y = 480
flag.pencolor('red')
flag.fillcolor('red')
flag.begin_fill()
flag.penup()
flag.forward(x/2)
flag.left(90)
flag.pendown()
flag.forward(y/2)
flag.left(90)
flag.forward(x/3)
flag.left(90)
flag.forward(y)
flag.left(90)
flag.forward(x/3)
flag.left(90)
flag.forward(y/2)
flag.end_fill()
# ô màu vàng ở giữa
#con trỏ hiện đang ở vị trí ở giữa cạnh phải
flag.penup()
flag.forward(y/2)
flag.left(90)
flag.forward(x/3)
flag.pendown()
flag.pencolor('yellow')
flag.fillcolor('yellow')
flag.begin_fill()
flag.forward(x/3)
flag.left(90)
flag.forward(y)
flag.left(90)
flag.forward(x/3)
flag.end_fill()
# ô màu xanh lá cây bên trái
flag.pencolor("green")
flag.penup()
flag.backward(x/3)
flag.pendown()
flag.backward(x/3)
flag.fillcolor('green')
flag.begin_fill()
flag.right(90)
flag.backward(y)
flag.right(90)
flag.backward(x/3)
flag.right(90)
flag.backward(y)
flag.end_fill()
# Vẽ ngôi sao chính giữa

flag.forward(y)
flag.right(90)
flag.penup()
flag.forward(x/6)
flag.right(90)
flag.forward(y/3)
flag.pendown
flag.fillcolor('green')
flag.begin_fill()
#cạnh 1
flag.right(18)
flag.forward(y/3)
#cạnh 2
flag.left(144)
flag.forward(y/3)
#cạnh 3
flag.left(144)
flag.forward(y/3)
#cạnh 4
flag.left(144)
flag.forward(y/3)
#cạnh 5
flag.left(144)
flag.forward(y/3)
flag.end_fill()



#hoàn thành
turtle.done()



```
