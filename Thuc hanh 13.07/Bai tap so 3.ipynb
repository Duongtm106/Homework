```python
#Vẽ 5 hình vuông cùng tâm đường chéo, mỗi hình vuông có cạnh song song cách nhau 40p
import turtle
t = turtle.Turtle()
n = 40
#cạnh hình vuông thứ nhất = 100
x = 100
t.pensize(2)
t.begin_fill()
t.color('green')
#Hình vuông thứ nhất màu đỏ (cạnh = 100)
t.penup()
t.forward(n/2)
t.pendown()
t.pencolor('red')
t.right(90)
t.forward(x/2)
t.right(90)
t.forward(x)
t.right(90)
t.forward(x)
t.right(90)
t.forward(x)
t.right(90)
t.forward(x/2)
#cạnh hình vuông thứ hai = 180 (x+2*n)
t.penup()
t.forward(x/2+n)
t.pendown()
t.pencolor('yellow')
t.right(90)
t.forward(x+n)
t.right(90)
t.forward(x+2*n)
t.right(90)
t.forward(x+2*n)
t.right(90)
t.forward(x+2*n)
t.right(90)
t.forward(n)
#cạnh hình vuông thứ 3 = 260 (x+4*n)
t.penup()
t.pencolor('blue')
t.forward(x+2*n)
t.pendown()
t.right(90)
t.forward(x+3*n)
t.right(90)
t.forward(x+4*n)
t.right(90)
t.forward(x+4*n)
t.right(90)
t.forward(x+4*n)
t.right(90)
t.forward(n)
#Cạnh hình vuông thứ 4 = 340 (x+ 6*n)
t.penup()
t.pencolor('green')
t.forward(x+4*n)
t.pendown()
t.right(90)
t.forward(x+5*n)
t.right(90)
t.forward(x+6*n)
t.right(90)
t.forward(x+6*n)
t.right(90)
t.forward(x+6*n)
t.right(90)
t.forward(n)
t.end_fill()
turtle.done()
```
