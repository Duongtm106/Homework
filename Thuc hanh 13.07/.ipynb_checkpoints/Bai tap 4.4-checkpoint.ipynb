```python
#Vẽ bát giác đều
import turtle
import math
t = turtle.Turtle()
#n là số tam giác tạo nên đa giác đều
n = 8
canh = 100
goc_tam = 360/n
goc_can = (180 - goc_tam)/2
canh_day = 2*canh*(math.sqrt(1 - math.sin(math.radians(goc_can))**2))
t.left(goc_tam/2)
#Vẽ tam giác số 1
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)    
#vẽ hình tam giác số 2
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)     
#vẽ hình tam giác số 3
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)    
#vẽ hình tam giác số 4
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)    
#vẽ hình tam giác số 5
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)    
#vẽ hình tam giác số 6
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)   
#vẽ hình tam giác số 7
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)
t.right(goc_tam*2)   
#vẽ hình tam giác số 8
t.forward(canh)
t.right(180-goc_can)
t.forward(canh_day)
t.right(180 - goc_can)
t.forward(canh)

turtle.done()
```
